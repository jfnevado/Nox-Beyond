```python
import requests
import numpy as np
import pandas as pd
from pycoingecko import CoinGeckoAPI 
from bs4 import BeautifulSoup
import lxml
import html5lib
from selenium import webdriver
from datetime import datetime, timedelta
```

```python
cg = CoinGeckoAPI()
avax_price = cg.get_coin_market_chart_by_id(id='avalanche-2', vs_currency='usd',  days='120')
```

```python
avax_price = avax_price['prices']
avax_price = pd.DataFrame(avax_price, columns=['timeStamp', 'price'])
avax_price['timeStamp'] = pd.to_datetime(avax_price['timeStamp'], unit='ms')
avax_price['timeStamp'] = avax_price['timeStamp'].dt.strftime('%b %d %Y')
avax_price
```

```python
avax_price = avax_price.groupby('timeStamp').mean()
avax_price
```

```python
total_balance = pd.read_csv('total_balance.csv')
total_balance
```

```python
avax_price['balance'] = round(avax_price['price'] * total_balance['total_avax'][0], 2)
```

```python
avax_price.to_csv("avax_chart_balance.csv", index=True)
```
